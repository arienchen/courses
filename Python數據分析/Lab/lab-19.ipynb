````python
# 簡單計數統計
df['收支科目'].value_counts()
````

## 新增自訂計算欄位

````python
# 新增一欄： 收支 = [收入,支出]
# 條件一： 收支科目 有 收入 或 支出 
# 例外：匿名捐贈, 金錢

to_del = df[df["收支科目"]=="金錢"]
to_del
````

````python

df.drop(to_del.index, inplace=True)
````

````python
# 條件二： 收支科目 為 xx支出 -> 支出，其他為 收入

# 方法一： apply + lambda
# 易讀， apply 功能彈性大
df['收支'] = df['收支科目'].apply(lambda x: '支出' if x.endswith('支出') else '收入')

# 方法二： NumPy.where
# 效能高，適合大型資料集
df['收支'] = np.where(df['收支科目'].str.endswith('支出'), '支出', '收入')

````

````python
df['收支'].value_counts()
````
